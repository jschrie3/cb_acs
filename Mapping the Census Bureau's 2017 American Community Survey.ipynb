# Mapping the 2017 American Community Survey

The American Community Survey is an annual survey maintained by the Census Burea. It is one of the oldest and most comprehensive surveys publicly available. In this notebook, County-level coordinates and data-points will be read in directly from census.gov and mapped. Community Action Agency labels will be added to the maps. Agency-level maps will also be created, but all of them will not be included here.

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import mapclassify as mc
import pysal
import re
import geoviews as gv
import geoviews.feature
from geoviews import dim, opts
from bokeh.plotting import figure
from bokeh.resources import CDN
from bokeh.embed import file_html
from bokeh.models import GeoJSONDataSource, LogColorMapper, LinearColorMapper
from bokeh.io import output_file, show
from bokeh.palettes import Viridis6 as palette
import json

gv.extension('bokeh')
```

```python
url = "https://www2.census.gov/geo/tiger/TIGER2017/TRACT/tl_2017_40_tract.zip"
tract = gpd.read_file(url)
tract.info()
```

```python
acs = pd.read_csv('ACS_17_5YR_S1701_with_ann.csv',
                  na_values=["-","*","**","(X)","N"])
acs.info()
```

```python
acs['GEOdisplaylabel'] = acs.loc[:, 'GEO.display-label']
acs['GEOID'] = acs['GEO.id2']

acs.info()
```

```python
y = acs.GEOdisplaylabel.str.split(',', expand=True)

y['y1'] = y.iloc[:,[1]]

y = y.y1.str.split(' ', expand=True)

y['y1'] = y.iloc[:,[1]]

y = y.y1

x = acs.join(y, how='left')

x.info()
```

```python
x['agencies'] = x.y1.replace(['Bryan',
                              'Carter',
                              'Coal',
                              'Love',
                              'Pontotoc'], 'Big 5') \
                    .replace(['Canadian',
                              'Oklahoma'], 'CAAOKC') \
                    .replace(['Beckham',
                              'Cotton',
                              'Jefferson',
                              'Kiowa',
                              'Roger',
                              'Tillman',
                              'Washita'], 'CADC') \
                    .replace(['Tulsa'], 'CAP Tulsa') \
                    .replace(['Mayes',
                              'Nowata',
                              'Rogers',
                              'Wagoner',
                              'Washington'], 'CARD') \
                    .replace(['Garfield',
                              'Grant'], 'CDSA') \
                    .replace(['Cleveland',
                              'Lincoln',
                              'Logan',
                              'Payne',
                              'Pottawatomie',
                              'Seminole'], 'COCAA') \
                    .replace(['Garvin',
                              'McClain',
                              'Stephens'], 'Delta') \
                    .replace(['Hughes',
                              'McIntosh',
                              'Okfuskee',
                              'Okmulgee'], 'DFCAF') \
                    .replace(['Comanche'], 'GPIF') \
                    .replace(['Atoka',
                              'Johnston',
                              'Marshall',
                              'Murray'], 'INCA') \
                    .replace(['Haskell',
                              'Latimer',
                              'Le',
                              'Muskogee',
                              'Pittsburg',
                              'Sequoyah'], 'KI BOIS') \
                    .replace(['Choctaw',
                              'McCurtain',
                              'Pushmataha'], 'Little Dixie') \
                    .replace(['Adair',
                              'Cherokee',
                              'Craig',
                              'Delaware',
                              'Ottawa'], 'NEOCAA') \
                    .replace(['Alfalfa',
                              'Beaver',
                              'Blaine',
                              'Cimarron',
                              'Custer',
                              'Dewey',
                              'Ellis',
                              'Harper',
                              'Kingfisher',
                              'Major',
                              'Texas',
                              'Woods',
                              'Woodward'], 'Opportunities') \
                    .replace(['Greer',
                              'Harmon',
                              'Jackson'], 'SOCAG') \
                    .replace(['Creek',
                              'Kay',
                              'Noble',
                              'Osage',
                              'Pawnee'], 'UCAP') \
                    .replace(['Caddo',
                              'Grady'], 'WVCAC')
```

```python
df = pd.merge(tract, 
             x,
             how='left',
             on='GEOID')
df.info()
```

```python
fig, ax = plt.subplots(figsize=(40, 40), subplot_kw={'aspect':'equal'})

df.plot(column='HC02_EST_VC01',
       cmap='YlGn',
       scheme='quantiles',
       legend=False,
       ax=ax,
       edgecolor='white',
       k=100,
       linewidth=0.125)

ax.set_axis_off()

plt.title("Number of People Below 100% of Poverty: ACS Census Tracts 2017", fontsize=40)

plt.show()
```

```python
okc = df[df.agencies == 'CAAOKC']
capt = df[df.agencies == 'CAP Tulsa']
```

```python
fig, ax = plt.subplots(figsize=(40, 40), subplot_kw={'aspect':'equal'})

okc.plot(column='HC02_EST_VC01',
       cmap='OrRd',
       scheme='quantiles',
       ax=ax,
       edgecolor='white',
       k=100,
       linewidth=0.125)

ax.set_axis_off()

plt.title("Number of People Below 100% of Poverty: CAA OKC Census Tracts 2017", fontsize=40)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(40, 40), subplot_kw={'aspect':'equal'})

capt.plot(column='HC02_EST_VC01',
       cmap='OrRd',
       scheme='quantiles',
       ax=ax,
       edgecolor='white',
       k=100,
       linewidth=0.125)

ax.set_axis_off()

plt.title("Number of People Below 100% of Poverty: CAP Tulsa Census Tracts 2017", fontsize=40)

plt.show()
```

```python
#Read data to json.
df_json = json.loads(df.to_json())

#Convert to String like object.
json_data = json.dumps(df_json)
```

```python
#Input GeoJSON source that contains features for plotting.
geo_source = GeoJSONDataSource(geojson = json_data)
```
